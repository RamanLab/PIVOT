### Upload file with aliqoute to sample mapping

```python
os.chdir(DATAPATH + "/GDC_{}/CNV/gistic_numbers".format(ctype))
fname = "aliquot.tsv"
data_aliq = pd.read_csv(fname, sep="\t", comment="#", header=0, index_col=9)
data_aliq.head()
```

```python
list(data_aliq.columns)
```

### Rename data columns
Column names changed to sample ID

```python
samp_map = {}
for x in data_cnv.columns:
    if x in list(data_aliq.index):
        if data_aliq.loc[x, "sample_submitter_id"] not in samp_map.values():
            samp_map[x] = data_aliq.loc[x, "sample_submitter_id"]
data_cnv.rename(columns=samp_map, inplace=True)
cols = [col for col in data_cnv.columns if col[:4] == "TCGA"]
data_cnv = data_cnv.loc[:, ["Gene Symbol", "Gene ID"] + cols]
```

```python
data_cnv.shape
```

## Covert the old gene symbols into relevant genes
Load hugo to ensemble id file and check for gene symbols not recognised.
Take unique list of genes and save to file.

```python
# Load hugo to ensemble id mapping data
os.chdir(DATAPATH)
fname = "hugo2ensmbl.txt"
data_map = pd.read_csv(fname, sep="\t", header=0)
data_map.drop_duplicates(["Ensembl gene ID"], inplace=True)
data_map.index = data_map["Ensembl gene ID"]
```

```python
data_map.head()
```

```python
data_cnv["Gene ID"] = [data_map.loc[ensg.split(".")[0], "Approved symbol"] if ensg.split(".")[0] in data_map.index else np.nan for ensg in data_cnv["Gene Symbol"]]
data_cnv.dropna(axis=0, subset=["Gene ID"], inplace=True)
# data_cnv.index = data_cnv["Gene ID"]
```

```python
data_cnv.head()
```

```python
%%time
samp_cols = [col for col in data_cnv.columns if col[:4] == "TCGA"]
data = [None] * len(samp_cols)
for idx, col in enumerate(samp_cols):
    temp = data_cnv[data_cnv[col] != 0]
    temp.loc[:, "Sample ID"] = [col] * len(temp)
    temp.loc[:, "Gene_symbol"] = temp.loc[:, "Gene ID"]
    temp.loc[:, "CNV"] = temp.loc[:, col]
    temp.index = ["{};{}".format(samp, gene) for samp, gene in zip(temp["Sample ID"], temp["Gene_symbol"])]
    data[idx] = pd.DataFrame(temp[["CNV"]], index=temp.index)
```

```python
%%time
data = pd.concat(data)
```

```python
data.shape
```

Save file

```python
os.chdir(PATH + "/data/GDC_{}/CNV".format(ctype))
fname="{}_cnv.tsv".format(ctype)
data.to_csv(fname, sep="\t", header=True, index=True)
```
